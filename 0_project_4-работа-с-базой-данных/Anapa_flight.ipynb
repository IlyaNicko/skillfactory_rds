```python
import pandas as pd
import matplotlib.pyplot as plt  
%matplotlib inline
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv('Anapa_winter_2017.csv')
df.head()
```

```python
df.info()
```

Имеются пропуски в столбцах ['boarded'] и ['total_cost']. Выведем рейсы с пропусками:

```python
df[df.boarded.isnull() == True]
```

Все рейсы  с пропусками представлены одним направлением Анапа - Новокузнецк. Т.к. у нас нет данных о количестве пассажиров и стоимости билетов по данному направлению придется исключить эти данные из датафрейма. 

```python
df = df.dropna()
```

```python
# преобразуем даты в правильный формат
df.actual_departure_local = df.actual_departure_local.apply(lambda x: pd.to_datetime(x))

# перевод длительности в int в часах
df.actual_duration = df.actual_duration.apply(lambda x: pd.to_timedelta(x[22:-11]).seconds/3600) 

# подсчет количества свободных мест (истинное и относительное)
df['free_seats'] = df['aircraft_seats'] - df['boarded']
df['free_seats_percent'] = (df['aircraft_seats'] - df['boarded'])/df['aircraft_seats']*100

# добавим названия дней недели
df['departure_dayname'] = df.actual_departure_local.apply(lambda x: x.day_name())
```

```python
# подсчет стоимости полета на основании расходов на топливо
# источники: Цена топлива - https://favt.gov.ru
#            Расход:  http://superjet.wikidot.com/wiki:rashod-topliva
#                     https://www.airlines-inform.ru/commercial-aircraft/boeing-737-300.html           

fuel_price = {12:38867 , 1:41435 , 2:39553} # Стоимость топлива дек2016, янв2017 и фев2017 
fuel_consumpt = {'Boeing 737-300': 2.6, 'Sukhoi Superjet-100':1.7 } # расход топлива, тонн/час

def flight_price(row):
    '''подсчет стоимости полета на основании расходов на топливо
       источники: Цена топлива: - https://favt.gov.ru
                  Расход:  - http://superjet.wikidot.com/wiki:rashod-topliva
                           - https://www.airlines-inform.ru/commercial-aircraft/boeing-737-300.html           
    '''
    
                                                                       
    price = (fuel_price[row['actual_departure_local'].month] 
    * fuel_consumpt[row['model']] 
    * row['actual_duration'])
    return price

df['flight_price'] = df.apply(flight_price,axis=1)   
```

```python
df['profit'] = df['total_cost'] - df['flight_price']
```

Выясним, куда зимой летают самолеты из Анапы?

```python
df['arrival_airport'].value_counts()
```

Получается из Анапы зимой можно улететь только в Москву или Белгород. 

```python
# разделим датасет по направлениям
df_M = df[df['arrival_airport']=='SVO'].sort_values(['actual_departure_local'])
df_B = df[df['arrival_airport']=='EGO'].sort_values(['actual_departure_local'])
```

```python
print(f'рейсы в Москву \n {df_M["flight_no"].value_counts()}')
print(f'рейсы в Белгород \n {df_B["flight_no"].value_counts()}')
```

Имеется одинаковое количество рейсов в Москву и Белгород. Судя по количеству рейсов они ежедневные, по одному на каждое направление. 

 Рассмотрим направления по отдельности
## Анапа - Москва

Выведем наиболее значимые статистические данные:

```python
for_count = ['boarded','free_seats','free_seats_percent','actual_duration','total_cost','flight_price','profit']
df_M[for_count].describe()
```

Средняя прибыль по направлению Анапа-Москва составляет 1473784 рублей, минимальная - 1203976 (ниже среднего на 18.3%)
Выведем рейсы, прибыль которых ниже среднего значения

```python
mean_less_M = df_M[df_M['profit']<df_M['profit'].mean()]
mean_less_M.sort_values(['profit']).reset_index()
```

Получается, что чуть боее половины рейсов приносит прибыль нижне среднего значения. 
Посмотрим, есть ли какие-либо закономерности в зависимости от дней месяца или недели.

```python
df_M = df_M.set_index('actual_departure_local')
dec = df_M[df_M.index.month==12]
jan = df_M[df_M.index.month==1]
feb = df_M[df_M.index.month==2]
```

```python
plt.figure(figsize=(15, 4))
plt.xticks(rotation=45)

plt.hlines(y=df_M['profit'].mean(), xmin = df_M.index[0],xmax = df_M.index[-1],color='red')
plt.vlines(x=df_M[df_M['departure_dayname']=='Sunday'].index,ymin=0, ymax=df_M['profit'].max(),color='gray')
plt.xlabel('actual_departure_local')
plt.ylabel('profit')
plt.bar(x=dec.index, height = dec.profit, color='c')
plt.bar(x=jan.index, height = jan.profit, color='y')
plt.bar(x=feb.index, height = feb.profit, color='pink')
```

Отмечается некоторая цикличность, однако связать ее с какими-либо частями месяца не получается. Посмотрим, что по дням недели

```python
mean_less_M['departure_dayname'].hist(bins=13)
```

Отмечается, что чаще всего низкая прибыль была на воскресный рейсах. Очевидно, что прибыль рейса зависит от количества пассажиров, но на всякий случай постороим матрицу корреляций.

```python
df_M[for_count].corr()
```

Видим сильную положительную корреляцию между прибылью и количеством пассажиров, а также сльную отрицательную корреляцию между прибылью и количеством свободных мест в самолете. Т.к. увеличить количество пассажиров мы не можем, стоит задуматься над уменьшением количеества свободных мест в самолете. Одим из вариантов можно предложить заменить Boeing 737-300(130мест) в воскресные даты на Sukhoi Superjet-100(97 мест).  

```python
plt.figure(figsize=(7, 7))
sns.boxplot(x = 'departure_dayname', y='boarded', data = df_M[df_M.departure_dayname=='Sunday'])
```

Очевидно, что количество пассажиров, летящих воскресными рейсами, больше, чем может поместиться на борт Sukhoi Superjet-100. Посчитаем, можно ли распределить оставшихся пассажиров на смежные даты, а именно на субботу и понедельник.

```python
df_M[df_M['departure_dayname']=='Sunday']['boarded'].mean() # среднее количество паасажиров в воскресенье
```

```python
df_M[df_M['departure_dayname']=='Sunday']['boarded'].sum() # улетело в воскресенье
```

```python
df_M[df_M['departure_dayname']=='Sunday']['boarded'].count()*97 # можено перевезти при замене самолета
```

```python
df_M.query("departure_dayname in['Saturday', 'Monday']")['boarded'].sum() # улетело в субботу и понедельник
```

```python
df_M.query("departure_dayname in['Saturday', 'Monday']")['free_seats'].mean() # среднее количество свободных мест
```

```python
df_M.query("departure_dayname in['Saturday', 'Monday']")['aircraft_seats'].sum() # может вылететь в субботу и понедельник
```

Таким образом общее количество пассажиров, которое нужно распределить на субботу и понедельник составляет 1379-1261=118 человек (в среднем 106-97=9 человек) , а свободных мест при этом  3380-2933=447 ( в среднем 17).
Т.к. Sukhoi Superjet-100 расходует топлива в 2.6/1.7=1.53 раза меньше, то выгода от замены самолета при условии сохранения стоимости билетов составит:

```python
(df_M[df_M['departure_dayname']=='Sunday']['flight_price'].sum()
-df_M[df_M['departure_dayname']=='Sunday']['flight_price'].sum()/1.53)
```

Или в относительных значениях:

```python
779351.6772549017/df_M['profit'].sum()*100
```

Столь низкий прирост прибыли можно объяснить тем, что стоимость топлива вносит не большой вклад в стоимость билета и составляет всего порядка 10%

```python
df_M.total_cost.sum()/df_M.flight_price.sum() # отношение цены билета к стоимости топлива
```

Однако снижение расхода топлива и уменьшение грузоподъемности самолета потенциально может привести к снижению аэропортовых и других сборов, тем самым дополнительно увеличивая прибыль. Помимо этого освободившийся Boeing 737-300 можно использовать на более загруженных трассах, что так же может привести к увеличению прибыли.
